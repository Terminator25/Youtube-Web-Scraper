```python
def scroll_to_bottom(driver):

    old_position = 0
    new_position = None

    while new_position != old_position:
        # Get old scroll position
        old_position = driver.execute_script(
                ("return (window.pageYOffset !== undefined) ?"
                 " window.pageYOffset : (document.documentElement ||"
                 " document.body.parentNode || document.body);"))
        # Sleep and Scroll
        time.sleep(1)
        driver.execute_script((
                "var scrollingElement = (document.scrollingElement ||"
                " document.body);scrollingElement.scrollTop ="
                " scrollingElement.scrollHeight;"))
        # Get new position
        new_position = driver.execute_script(
                ("return (window.pageYOffset !== undefined) ?"
                 " window.pageYOffset : (document.documentElement ||"
                 " document.body.parentNode || document.body);"))
```

```python
read=webdriver.Chrome(ChromeDriverManager().install())

url="https://blog.feedspot.com/technology_youtube_channels"
read.get(url)
time.sleep(3)
channels=read.find_elements_by_class_name('tlink')
channellink=read.find_elements_by_class_name('ext')
```

```python
channel_name=[]
channel_link=[]
```

```python
for i in range(0,20):
    channel_name.append(channels[i].text)
    channel_link.append(channellink[i].get_attribute('href'))
```

```python
def scroll_to_bottom(driver):

    old_position = 0
    new_position = None
    timeout = 60   # [seconds]
    timeout_start = time.time()

    while time.time() < timeout_start + timeout:
        test = 0
        if test == 5:
            break
        test -= 1
        # Get old scroll position
        old_position = driver.execute_script(
                ("return (window.pageYOffset !== undefined) ?"
                 " window.pageYOffset : (document.documentElement ||"
                 " document.body.parentNode || document.body);"))
        # Sleep and Scroll
        time.sleep(1)
        driver.execute_script((
                "var scrollingElement = (document.scrollingElement ||"
                " document.body);scrollingElement.scrollTop ="
                " scrollingElement.scrollHeight;"))
        # Get new position
        new_position = driver.execute_script(
                ("return (window.pageYOffset !== undefined) ?"
                 " window.pageYOffset : (document.documentElement ||"
                 " document.body.parentNode || document.body);"))
```

```python
read=webdriver.Chrome(ChromeDriverManager().install())
for i in range(1,20):
    read.get(channel_link[i])
    scroll_to_bottom(read)
    videos=read.find_elements_by_xpath('//*[@id="video-title"]')
    links=[]
    video_names=[]
    for video in videos:
        links.append(video.get_property('href'))
        video_names.append(video.text)
    creator={'Channel Name':channel_name[i], 'Video Title':video_names, 'Link':links }
    channel_data=pd.DataFrame(creator)
    name= 'channeldata'+str(i)
    channel_data.to_csv(name)
```
